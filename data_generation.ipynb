```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta
from geopy.distance import geodesic
```

```python
# Initialize Faker for generating fake customer names
fake = Faker()

# Create an empty DataFrame
data = pd.DataFrame(columns=[
    'OrderID', 'CustomerID', 'OrderDate', 'DeliveryTime(min)',
    'TotalOrderValue', 'ItemsOrdered', 'DeliveryLocation(Latitude)', 'DeliveryLocation(Longitude)',
    'DispatchLocation(Latitude)', 'DispatchLocation(Longitude)', 'DistanceFromRestaurant(KM)'
])

# Define the number of records you want in your dataset
num_records = 500
```

```python

# Define the date range for generating orders
start_date = datetime(2023, 7, 1)
end_date = datetime(2023, 11, 5)
```

```python
# Mumbai coordinates
mumbai_latitude = 19.0760
mumbai_longitude = 72.8777

# Tilak Nagar Lok Terminal coordinates
tilak_nagar_latitude = 19.0047
tilak_nagar_longitude = 72.8415


# Create a dictionary to map items to their constant prices
item_prices = {
    'Pizza': 150,
    'Burgers': 75,
    'Sushi': 200,
    'Pasta': 180,
    'Sandwiches': 75
}
```

```python
# Generate random data for the specified number of records
for order_id in range(1, num_records + 1):
  # Define the range for average speed (in km/h)
    average_speed = random.randint(45, 80)
    customer_id = f'Cust{order_id:03}'
    order_date = fake.date_time_between_dates(start_date, end_date)


    # Generate a random number of items in each order (between 1 and 5)
    num_items = random.randint(1, 5)
    items_ordered = [fake.random_element(elements=list(item_prices.keys())) for _ in range(num_items)]

    # Calculate the total order value for all items in the order
    total_order_value = sum(item_prices[item] for item in items_ordered)

    # Generate coordinates around Mumbai
    customer_latitude = round(tilak_nagar_latitude + random.uniform(-0.5, 0.5), 4)
    customer_longitude = round(tilak_nagar_longitude + random.uniform(-0.5, 0.5), 4)

    distance_km = geodesic(
        (tilak_nagar_latitude, tilak_nagar_longitude),
        (customer_latitude, customer_longitude)
    ).kilometers

    delivery_time = 60*(distance_km/average_speed)
    print(delivery_time)

    data = data.append({
        'OrderID': order_id,
        'CustomerID': customer_id,
        'OrderDate': order_date.date(),
        'DeliveryTime(min)': delivery_time,
        'TotalOrderValue': total_order_value,
        'ItemsOrdered': ', '.join(items_ordered),
        'DeliveryLocation(Latitude)': customer_latitude,
        'DeliveryLocation(Longitude)': customer_longitude,
        'DispatchLocation(Latitude)': tilak_nagar_latitude,
        'DispatchLocation(Longitude)': tilak_nagar_longitude,
        'DistanceFromRestaurant(KM)': distance_km
    }, ignore_index=True)

# Sort the DataFrame by date and time in ascending order
# data['DeliveryTime'] = pd.to_timedelta(data['DeliveryTime'], unit='m')  # Convert to Timedelta
data = data.sort_values(by='OrderDate')

# Save the generated data to a CSV file

```
